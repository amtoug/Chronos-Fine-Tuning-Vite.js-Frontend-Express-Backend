```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("eco2mix_Data.csv",index_col="Unnamed: 0")
df
```

```python
df_train=df[df['Date - Heure']<"2025-10-01"]
df_train
```

```python
df_test=df[df['Date - Heure']>"2025-10-01"]
df_test.head()
```

```python
import torch
from chronos import ChronosPipeline
import numpy as np
```

```python
pipeline = ChronosPipeline.from_pretrained(
    #Modele par defaut
  "amazon/chronos-t5-tiny",
  device_map="cuda",
  torch_dtype=torch.bfloat16,
)

# context must be either a 1D tensor, a list of 1D tensors,
# or a left-padded 2D tensor with batch as the first dimension
forecast = pipeline.predict(torch.tensor(df_train["target"]),
                            prediction_length = 48,
                            num_samples=20)  # shape [num_series, num_samples, prediction_length]
```

```python
forecast[0].shape
```

```python
forecast = pipeline.predict(torch.tensor(df_test["target"].values[0:512]),
                            prediction_length = 48,
                            num_samples=20)

low, median, high = np.quantile(forecast[0].numpy(), [0.1, 0.5, 0.9], axis=0)

plt.figure(figsize=(8, 4))
plt.plot(df_test["target"].values[512:512+48], color="royalblue", label="historical data")
#median === prediction
plt.plot(median, color="tomato", label="median forecast")
plt.fill_between(list(range(48)), low, high, color="tomato", alpha=0.3, label="80% prediction interval")
plt.legend()
plt.grid()
plt.show()
```

```python
#prediction des 4 jours suivents 22*24=528
# 512 c'est le context par defaut 21 jour+8h
context_window=512
i=7

forecast = pipeline.predict(torch.tensor(df_test["target"].values[i*24:context_window+i*24]),
                            prediction_length = 24,
                            num_samples=20)

low, median, high = np.quantile(forecast[0].numpy(), [0.1, 0.5, 0.9], axis=0)

plt.figure(figsize=(8, 4))
plt.plot(df_test["target"].values[context_window+i*24:context_window+(i+1)*24], color="royalblue", label="historical data")
#Y
plt.plot(median, color="tomato", label="median forecast")
#X
plt.fill_between(list(range(24)), low, high, color="tomato", alpha=0.3, label="80% prediction interval")
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_percentage_error
#Nbr des jours de prediction
Steps=int((len(df_test["target"].values)-context_window)/24)
#Liste des err pour chaque prediction
Err=[]
for i in range(Steps):

  True_value=df_test["target"].values[context_window+i*24:context_window+(i+1)*24]

  if np.any(np.isnan(True_value)):
    continue
  forecast = pipeline.predict(torch.tensor(df_test["target"].values[i*24:context_window+i*24]),
                            prediction_length = 24,
                            num_samples=20)
  median = np.quantile(forecast[0].numpy(), [0.5], axis=0)
  Err.append(mean_absolute_percentage_error(True_value,median[0]))#(valeur,prediction)
print(Err)
```

```python
np.mean(Err)
#alors 2.28% des errurs(mean_absolute_percentage_error)
```

# **Test de stationnarites**

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(df_train["target"])

print('ADF Statistic:', result[0])
print('p-value:', result[1])
print('Critical Values:', result[4])

if result[1] < 0.05:
    print("La sÃ©rie est probablement stationnaire")
else:
    print("La sÃ©rie n'est probablement pas stationnaire")
```
